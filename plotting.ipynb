```python
import matplotlib.pyplot as plt
import numpy as np

##Plotting av funksjoner:
def f(x_f):
    return np.e**(-x_f/4)*np.arctan(x_f) #funksjonen til f
def g(x_g):
    return np.arctan(x_g)-(4/(x_g**2+1)) #funksjonen for å finne maksimalpunkt
x = np.linspace(-10, 10, 400)  #x verdier, 400 punkter
plt.plot(x, f(x), label="f(x) = e^(-x/4)*arctan(x)") #plot for funksjonen f
plt.plot(x, g(x), label="g(x) = arctan(x)-(4/(x^2+1))") #plot for funksjonen g,som er likningen for toppunktet til f
#der hvor g er lik 0 er også topppunktet til f.
plt.grid(True) #for å vise noen linjer langs verdier for å lettere se hva et punkt sine verdi er
plt.legend() #for å differensiere mellom funksjonene
plt.show() #viser plottet
```
