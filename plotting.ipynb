```python
import numpy as np

##Newtons Metode
#Se jpg bevis_for_df_lik_null og beregninger_newtonsmetode for hvordan jeg kom fram til forskjellige verdier brukt i oppgaven.
x = 1.5 #x_0
pres = 1e-5 #presisjon
x_temp = 100 #lager en midlertidig x-verdi

while abs(x-x_temp) > pres:
    x_temp = x #kopi av den gamle x
    x = x - (np.arctan(x) - (4/(x**2+1))) / ( (1/(x**2+1)) + (8*x/(x**2+1)**2) ) #her er f'(x) = 0 uttrykket og

print(f'x = {x:.4f}') #svaret
```

```python
import matplotlib.pyplot as plt
import numpy as np

##For å vise at e^(-x/4) aldri treffer null.
def e(x_e):
    return np.e**(-x_e/4)

##Plotting av funksjonen for å vise at den aldri treffer null:
x = np.linspace(-10, 100, 400)
plt.plot(x,e(x),label="e^(-x/4)")
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

##Plotting av funksjoner:
def f(x_f):
    return np.e**(-x_f/4)*np.arctan(x_f) #funksjonen til f
def g(x_g):
    return np.arctan(x_g)-(4/(x_g**2+1)) #funksjonen for å finne maksimalpunkt

x = np.linspace(-10, 10, 400)  #x verdier, 400 punkter
plt.plot(x, f(x), label="f(x) = e^(-x/4)*arctan(x)") #plot for funksjonen f
plt.plot(x, g(x), label="g(x) = arctan(x)-(4/(x^2+1))") #plot for funksjonen g,som er likningen for toppunktet til f
#der hvor g er lik 0 er også topppunktet til f.
plt.grid(True) #for å vise noen linjer langs verdier for å lettere se hva et punkt sine verdi er
plt.legend() #for å differensiere mellom funksjonene
plt.show() #viser plottet
```
