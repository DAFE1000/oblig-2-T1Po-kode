```python
import matplotlib.pyplot as plt
import numpy as np

##Plotting av funksjoner:
def f(x):
    return np.e**(-x/4)*np.arctan(x)
x = np.linspace(-10, 10, 400)  #x verdier, 400 punkter
plt.plot(x, f(x)) #lag et plot og plott verdiene
plt.grid(True) #for å vise noen linjer langs verdier for å lettere se hva et punkt sin verdi er
plt.show() #viser plottet
```
