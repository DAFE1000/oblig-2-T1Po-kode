```python
##nødvendige imports
import numpy as np
import matplotlib.pyplot as plt
###Funksjoner som blir brukt i andre celler
#funksjonen til f
def f(x_f):
    return np.e**(-x_f/4)*np.arctan(x_f)
#funksjonen for å finne maksimalpunkt også kjent som, f'(x) = 0
def g(x_g):
    return np.arctan(x_g)-(4/(x_g**2+1))
#funksjon for å vise at e^(-x/4) aldri treffer null.
def e(x_e):
    return np.e**(-x_e/4)
```

```python
##Newtons Metode
#Se jpg bevis_for_df_lik_null og beregninger_newtonsmetode for hvordan jeg kom fram til forskjellige verdier brukt i oppgaven.
nm_x = 1.5 #x_0
pres = 1e-5 #presisjon
nm_x_temp = 100 #lager en midlertidig x-verdi

while abs(nm_x-nm_x_temp) > pres:
    nm_x_temp = nm_x #kopi av den gamle x
    nm_x = nm_x - (np.arctan(nm_x) - (4/(nm_x**2+1))) / ( (1/(nm_x**2+1)) + (8*nm_x/(nm_x**2+1)**2) ) #her er f'(x) = 0 uttrykket og den deriverte til det uttrykket brukt i newtons metode.

print(f'x = {nm_x:.4f}') #svaret
print(f'x = {f(nm_x):.4f}') #y-verdien i f
```

```python
##Plotting av funksjonen e for å vise at den aldri treffer null:
x = np.linspace(-10, 100, 400)
plt.plot(x,e(x),label="e^(-x/4)")
plt.grid()
plt.legend()
plt.show()
```

```python
##Plotting av funksjonene f og g + toppunktet til f:
x = np.linspace(-10, 10, 400)  #x verdier, 400 punkter
plt.plot(x, f(x), label="f(x) = e^(-x/4)*arctan(x)") #plot for funksjonen f
plt.plot(nm_x, f(nm_x), 'ro', label="toppunkt til f(x)") #plot for toppunktet til f
plt.annotate(f"({nm_x:.4f}, {f(nm_x):.4f})", xy=(nm_x, f(nm_x)), textcoords="offset points", xytext=(5, 5)) #Dette er bare for å vise hva koodinatene til toppunktet er.
plt.plot(x, g(x), label="g(x) = arctan(x)-(4/(x^2+1))") #plot for funksjonen g, som er likningen for toppunktet til f, der hvor g er lik 0 er også toppunktet til f.
plt.grid(True) #for å vise noen linjer langs verdier for å lettere se hva et punkt sine verdi er
plt.legend() #for å differensiere mellom funksjonene
plt.show() #viser plottet
```
